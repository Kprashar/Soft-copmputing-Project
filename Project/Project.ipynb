```python
import numpy as np
import skfuzzy as fuzzy
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt
import pandas as pd
```

All the necessary liberaries are imported

```python
error=ctrl.Antecedent(np.arange(-1,1,0.001),'error')
```

First antecedent created with its Universal range 

```python
error['NBe']=fuzzy.trimf(error.universe,[-1,-1,-0.5])
error['NKe']=fuzzy.trimf(error.universe,[-1,-0.5,0])
error['Se']=fuzzy.trimf(error.universe,[-0.5,0,0.5])
error['PKe']=fuzzy.trimf(error.universe,[0,0.5,1])
error['PBe']=fuzzy.trimf(error.universe,[0.5,1,1])
```

All the subparts with membership function

```python
change_error=ctrl.Antecedent(np.arange(-1,1,0.001),'change_error')
```

Second antecedent created with its Universal range 

```python
change_error['NBe']=fuzzy.trimf(change_error.universe,[-1,-1,-0.5])
change_error['NKe']=fuzzy.trimf(change_error.universe,[-1,-0.5,0])
change_error['Se']=fuzzy.trimf(change_error.universe,[-0.5,0,0.5])
change_error['PKe']=fuzzy.trimf(change_error.universe,[0,0.5,1])
change_error['PBe']=fuzzy.trimf(change_error.universe,[0.5,1,1])
```

All the subparts with membership function

```python
result=ctrl.Consequent(np.arange(-1,1,0.001),'result')
```

Consequent created with its Universal range

```python
result['NBe']=fuzzy.trimf(result.universe,[-1,-1,-0.5])
result['NKe']=fuzzy.trimf(result.universe,[-1,-0.5,0])
result['Se']=fuzzy.trimf(result.universe,[-0.5,0,0.5])
result['PKe']=fuzzy.trimf(result.universe,[0,0.5,1])
result['PBe']=fuzzy.trimf(result.universe,[0.5,1,1])
```

All the membership function

```python
rule1=ctrl.Rule(error['NBe'] &change_error['NBe'],result['NBe'])
rule2=ctrl.Rule(error['NKe'] &change_error['NBe'],result['NBe'])
rule3=ctrl.Rule(error['Se'] &change_error['NBe'],result['NKe'])
rule4=ctrl.Rule(error['PKe'] &change_error['NBe'],result['NKe'])
rule5=ctrl.Rule(error['PBe'] &change_error['NBe'],result['Se'])
rule6=ctrl.Rule(error['NBe'] &change_error['NKe'],result['NBe'])
rule7=ctrl.Rule(error['NKe'] &change_error['NKe'],result['NKe'])
rule8=ctrl.Rule(error['Se'] &change_error['NKe'],result['NKe'])
rule9=ctrl.Rule(error['PKe'] &change_error['NKe'],result['Se'])
rule10=ctrl.Rule(error['PBe'] &change_error['NKe'],result['PKe'])
rule11=ctrl.Rule(error['NBe'] &change_error['Se'],result['NKe'])
rule12=ctrl.Rule(error['NKe'] &change_error['Se'],result['NKe'])
rule13=ctrl.Rule(error['Se'] &change_error['Se'],result['Se'])
rule14=ctrl.Rule(error['PKe'] &change_error['Se'],result['PKe'])
rule15=ctrl.Rule(error['PBe'] &change_error['Se'],result['PKe'])
rule16=ctrl.Rule(error['NBe'] &change_error['PKe'],result['NKe'])
rule17=ctrl.Rule(error['NKe'] &change_error['PKe'],result['Se'])
rule18=ctrl.Rule(error['Se'] &change_error['PKe'],result['PKe'])
rule19=ctrl.Rule(error['PKe'] &change_error['PKe'],result['PKe'])
rule20=ctrl.Rule(error['PBe'] &change_error['PKe'],result['PBe'])
rule21=ctrl.Rule(error['NBe'] &change_error['PBe'],result['Se'])
rule22=ctrl.Rule(error['NKe'] &change_error['PBe'],result['PKe'])
rule23=ctrl.Rule(error['Se'] &change_error['PBe'],result['PKe'])
rule24=ctrl.Rule(error['PKe'] &change_error['PBe'],result['PBe'])
rule25=ctrl.Rule(error['PBe'] &change_error['PBe'],result['PBe'])
```

Allt he 25 rule to calculate consquent

```python
tipping=ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10,rule11,rule12,rule13,rule14,rule15,rule16,rule17,rule18,rule19,rule20,rule21,rule22,rule23,rule24,rule25])
Tip=ctrl.ControlSystemSimulation(tipping)
```

Puting the rule into the system

```python
error_list=[]
change_error_list=[]
result_list=[]
y=np.arange(1,20,1)
```

Declaring some list for futher use

```python
df=pd.read_csv("data.csv")
df=df.values
```

Opening csv file

```python
i=0
while(i<19):
    Tip.input['error']=df[i][0]
    error_list.append(df[i][0])
    Tip.input['change_error']=df[i][1]
    change_error_list.append(df[i][1])
    Tip.compute()
    print(Tip.output['result'])
    result_list.append(Tip.output['result'])
    i=i+1
```
